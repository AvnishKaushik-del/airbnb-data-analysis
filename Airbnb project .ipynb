```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("compressed_data.csv.gz")
```

```python
df.head()
```

```python
df.columns
```

# Checking missing valuse

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling missing value

```python
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0,'last review' : df['last review'].min()},inplace = True)
```

```python
df.dropna(subset = ['NAME','host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["licenese","house_rules"],errors='ignore')
```

```python
df.head()
```

```python
#Remove dollar signs and convert to float.
df['price']=df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace= True)
```

```python
df.info()
```

# descriptive statistics

```python
df.describe()
```

# Visualization

# What is the distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins=50,kde=True,color='red')
plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

```python
#The histogram  shows a fairly even distribution of listing prices across different price ranges,indicating no partivular concentration of listings in any specific price range.
#The KDE line helps visualize this even spread more clearly,confirming that the dataset contains listings witha wide variety of prices.
```

# How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type",data=df,color="yellow")
plt.title("room type distribution")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

# How are listings distributed across different neighborhoods?

```python
plt.figure(figsize = (12,8))
sns.countplot(y= "neighbourhood group" ,data=df, color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title('Number of Listings by Neighbouhood Group')
plt.xlabel("Count")
plt.ylabel('Neighborhood Group')
plt.show()
```

# What is relationship between price and room type?

```python
plt.figure(figsize = (10,6))
sns.boxplot(x= 'room type' ,y='price',hue='room type',data=df,palette='Set1')
plt.title('Price vs. Room Type')

plt.xlabel('Room Type')
plt.ylabel("Price ($)")
plt.legend(title='Room Type')
plt.show()
```

# How has the  number of reviews change over with time?

```python
df['last review']=pd.to_datetime(df['last review'])
reviews_over_time=df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
